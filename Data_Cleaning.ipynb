# Cleaning Process

### Importing Necessary Libraries

```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sb
```

### Reading the data

```python
pd.set_option("display.max_columns",None)
data=pd.read_csv("data.csv")
data
```

### Works to be done in Cleaning process
- **Drop ClaimNumber** Column as it is not useful for the prediction.
- Remove **Time** values from **T** in **DateTimeOfAccident** and **DateReported**.
- Change Datatype of **DateTimeOfAccident** and **DateReported** to **datetime64**.
- Count the **Total.No.Of.Days** between **DateTimeOfAccident** and **DateReported** and make a **newcolumn** with datatype as **int**.

```python
data.shape
```

```python
data.dtypes
```

**Droping** the column **ClaimNumber**

```python
data.drop("ClaimNumber",1,inplace=True)
data
```

#### Checking the null values and filling with the necessary values

```python
data.isnull().sum()
```

Checking the count of **Unknown** values in the **MaritalStatus**.

```python
data["MaritalStatus"].value_counts()
```

Here, We are assigning the value **U (unknown)** to the nullvalue rows of **MaritalStatus**.

```python
data1=data.copy()
data1["MaritalStatus"]=data1["MaritalStatus"].fillna("U")
data1["MaritalStatus"].value_counts()
```

All the **Null Values** are cleared.

```python
data1.isnull().sum()
```

Checking for **Dump Values** in the **Gender** columns as it is useless for the prediction.

```python
data1["Gender"].value_counts()
```

As, We got **2 U(Unknown)values**, We are deleting those 2 rows by getting their **index**. 
And also **Reindexing** the **DataFrame**.

```python
i=data1[data1["Gender"]=="U"].index.to_list()
print(i)
print(data1.shape)
data1.drop(i,axis=0,inplace=True)
data1.reset_index(drop = True,inplace=True)
print(data1.shape)
```

This is to check the **ValueCounts** of few columns in the **DataFrame** to see any **irrelavent values** are present or not.

```python
def valuecounts(a):
    print(data1[a].value_counts())
    
valuecounts("Gender")
valuecounts("PartTimeFullTime")
valuecounts("DaysWorkedPerWeek")
valuecounts("DependentChildren")
valuecounts("DependentsOther")
```

To check whether **Zero (0)** is present in the age column as it is useless.

```python
data1[data1["Age"]==0]["Age"].count()
```

Removing **Unnecessory Strings** to get **Actual Date Format**.

```python
df=data1.copy()
df["DateTimeOfAccident"]=df["DateTimeOfAccident"].map(lambda x: str(x)[:-10])
df["DateReported"]=df["DateReported"].map(lambda x: str(x)[0:10])
```

Changing the **Data Type** of **"DateTimeOfAccident"** and **"DateReported"** to **datetime64**.

```python
df["DateTimeOfAccident"]=pd.to_datetime(df["DateTimeOfAccident"])
df["DateReported"]=pd.to_datetime(df["DateReported"])
df
```

Creating a new column **DaysGap** by **Subtracting** the **Dates**.

```python
df1=df.copy()
df1["DaysGap"]=df1["DateReported"]-df1["DateTimeOfAccident"]
```

**Droping** **"DateTimeOfAccident"** and **"DateReported"**.

```python
df1.drop("DateTimeOfAccident",axis=1,inplace=True)
df1.drop("DateReported",axis=1,inplace=True)
```

To change the datatype of the **DaysGap** column to **int64**:
- First Change it to **Object** datatype.
- To **Delete** the **Unwanted Content** in the column.
- Then change the datatype into **int64**.

```python
df1["DaysGap"]=df1["DaysGap"].astype(object)
df1['DaysGap'].dtype
```

```python
df1["DaysGap"]=df1["DaysGap"].apply(str).str.replace(" days 00:00:00","")
df1["DaysGap"]=df1["DaysGap"].astype(int)
df1['DaysGap'].dtypes
```

Here, Few Rows are having **Zero(0)** in  **"HoursWorkedPerWeek"** column which is useless for the Prediction.
- So , **Droping** those Rows , And Reindexing the **DataFrame**.

```python
dataset=df1.copy()
print(len(dataset[dataset["HoursWorkedPerWeek"]==0]))
print(dataset.shape)
i=dataset[dataset["HoursWorkedPerWeek"]==0].index.to_list()
dataset.drop(i,0,inplace=True)
dataset.reset_index(drop = True,inplace=True)
dataset.shape
```

Deriving **3 New-Columns** with the help of **Existing** Columns.
- **"WorkingHoursPerDay"** is derived from **"HoursWorkedPerWeek"** and **"DaysWorkedPerWeek"**.
- **"WagesPerDay"** is derived from **"WeeklyWages"** and **"DaysWorkedPerWeek"**.
- **"WagesPerHour"** is derived from **"WeeklyWages"** and **"HoursWorkedPerWeek"**.

```python
dataset1=dataset.copy()
dataset1["WorkingHoursPerDay"]=dataset1["HoursWorkedPerWeek"]/dataset1["DaysWorkedPerWeek"]
dataset1["WagesPerDay"]=dataset1["WeeklyWages"]/dataset1["DaysWorkedPerWeek"]
dataset1["WagesPerHour"]=dataset1["WeeklyWages"]/dataset1["HoursWorkedPerWeek"]
dataset1
```

**Rearranging** the Columns.

```python
dataset1=dataset1.reindex(columns=["Age","Gender","MaritalStatus","DependentChildren","DependentsOther","PartTimeFullTime","WorkingHoursPerDay","WagesPerHour","WagesPerDay","WeeklyWages","HoursWorkedPerWeek","DaysWorkedPerWeek","DaysGap","ClaimDescriptionKeyword_0","ClaimDescriptionKeyword_1","ClaimDescriptionKeyword_2","ClaimDescriptionKeyword_3","ClaimDescriptionKeyword_4","ClaimDescriptionKeyword_5","ClaimDescriptionKeyword_6","ClaimDescriptionKeyword_7","ClaimDescriptionKeyword_8","ClaimDescriptionKeyword_9","ClaimDescriptionKeyword_10","ClaimDescriptionKeyword_11","InitialIncurredClaimsCost","UltimateIncurredClaimCost"])
dataset1
```

**Creating** a new **Data File** for further processing.

```python
dataset1.to_csv("cleaned_data.csv",index=False)
```
