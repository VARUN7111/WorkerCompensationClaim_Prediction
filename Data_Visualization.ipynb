```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("cleaned_data.csv")
pd.set_option("display.max_columns",None)
data
```

```python
data.dtypes
```

```python
# heatmap
plt.subplots(figsize=(20,10))
sns.heatmap(data.corr(),cmap="Greens",annot=True)
```

```python
#pie chart

labels = ["MALE","FEMALE"]
gender = [ len(data[data["Gender"]=="M"].index.to_list()) , len(data[data["Gender"]=="F"].index.to_list())]
#colours
colours = [ "#ff9999" , "#66b3ff"]
#explosion
explode = [0.05,0.05]
fig1,ax1 = plt.subplots()
plt.pie(gender , colors=colours , labels=labels , autopct="%1.1f%%",wedgeprops = {'linewidth': 1},startangle=60,pctdistance=0.85)
#draw circle
centre_circle = plt.Circle((0,0),0.70,fc="white")
fig=plt.gcf()
fig.gca().add_artist(centre_circle)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis("equal")
ax1.set_title("Pie Chart - MALE vs FEMALE" , fontsize=13,loc="right")
plt.tight_layout()
plt.legend()
plt.show()
```

```python
#pie chart

labels = ["FULLTIME","PARTTIME"]
gender = [ len(data[data["PartTimeFullTime"]=="F"].index.to_list()) , len(data[data["PartTimeFullTime"]=="P"].index.to_list())]
#colours
colours = [ "#ff6666" , "#66b3ff" ]
#explosion
explode = [0.05,0.05]
fig1,ax1 = plt.subplots()
plt.pie(gender , colors=colours , labels=labels,explode=explode, autopct="%1.1f%%",wedgeprops = {'linewidth': 1},startangle=30,pctdistance=0.85)
#draw circle
centre_circle = plt.Circle((0,0),0.70,fc="white")
fig=plt.gcf()
fig.gca().add_artist(centre_circle)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis("equal")
ax1.set_title("Pie Chart - FULLTIME vs PARTTIME" , fontsize=13,loc="right")
plt.tight_layout()
plt.legend()
plt.show()
```

```python
#pie chart

labels = ["MARRIED","SINGLE","UNKNOWN"]
gender = [ len(data[data["MaritalStatus"]=="M"].index.to_list()) , len(data[data["MaritalStatus"]=="S"].index.to_list()) , len(data[data["MaritalStatus"]=="U"].index.to_list())]
#colours
colours = [ "#ff9999" , "#66b3ff" , "#99ff99"]
#explosion
explode = [0,0.05,0.05]
fig1,ax1 = plt.subplots()
plt.pie(gender , colors=colours , labels=labels , autopct="%1.1f%%",explode=explode,startangle=30,pctdistance=0.85)
#draw circle
centre_circle = plt.Circle((0,0),0.70,fc="white")
fig=plt.gcf()
fig.gca().add_artist(centre_circle)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis("equal")
ax1.set_title("Pie Chart - Marital Status" , fontsize=13,loc="right")
plt.tight_layout()
plt.legend()
plt.show()
```
